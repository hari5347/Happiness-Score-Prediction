```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../Datasets/Merged.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.replace({"..":np.nan},inplace=True)
```

```python
df.head()
```

```python
threshold = df.shape[0] * .50
df = df.dropna(thresh=threshold,axis=1,)
```

```python
df.info()
```

```python
plt.figure(figsize=(20,10))
ax=sns.heatmap(df.isnull(),yticklabels=100,cbar=False,cmap='viridis')
```

```python
columns = df.columns.drop(['Country Name','Region', 'Income Group', 'Lending category', 'Happiness Score', 'Year'])
cols = df.columns.drop(['Country Name','Happiness Score'])
```

```python
df[columns]= df[columns].apply(pd.to_numeric)
```

```python
df.drop(columns='Lending category',inplace=True)
```

```python
df.head()
```

```python
columns
```

```python
df[columns] = df.groupby(['Year','Region','Income Group'])[columns].transform(lambda x: x.fillna(x.mean()))
```
