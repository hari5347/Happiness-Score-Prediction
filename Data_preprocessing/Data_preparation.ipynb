```python
df.dropna(how='all',inplace=True)
```

```python
df.dropna(subset=['Country Code'],inplace=True)
```

```python
df.tail()
```

```python
df_2015 = df.drop(['2016 [YR2016]','2017 [YR2017]','Country Code'],axis=1)
df_2016 = df.drop(['2015 [YR2015]','2017 [YR2017]','Country Code'],axis=1)
df_2017 = df.drop(['2015 [YR2015]','2016 [YR2016]','Country Code'],axis=1)
```

```python
df_2015 = df_2015.groupby('Country Name').apply(lambda x:x.iloc[:,1:].T)
df_2016 = df_2016.groupby('Country Name').apply(lambda x:x.iloc[:,1:].T)
df_2017 = df_2017.groupby('Country Name').apply(lambda x:x.iloc[:,1:].T)
```

```python
df_2015.head()
```

```python
df_2015 = df_2015.reset_index()
df_2016 = df_2016.reset_index()
df_2017 = df_2017.reset_index()
```

```python
df_2015.head()
```

```python
df_2015.drop(columns='level_1',axis=1,inplace=True)
df_2016.drop(columns='level_1',axis=1,inplace=True)
df_2017.drop(columns='level_1',axis=1,inplace=True)
```

```python
columns = df_2015.iloc[0,1:].values
columns = [i for i in columns]
columns.insert(0,'Country Name')
```

```python
df_2015=df_2015.iloc[[i for i in range(1,df_2015.shape[0],2)],:]
df_2016=df_2016.iloc[[i for i in range(1,df_2016.shape[0],2)],:]
df_2017=df_2017.iloc[[i for i in range(1,df_2017.shape[0],2)],:]
```

```python
df_2015.head()
```

```python
df_2015 = df_2015.reset_index()
df_2016 = df_2016.reset_index()
df_2017 = df_2017.reset_index()
```

```python
df_2015.drop(columns='index',inplace=True)
df_2016.drop(columns='index',inplace=True)
df_2017.drop(columns='index',inplace=True)
```

```python
df_2015.columns = columns
df_2016.columns = columns
df_2017.columns = columns
```

```python
df_2015.head()
```

```python
Year = pd.Series([2015 for i in range(df_2015.shape[0])],name='Year')
df_2015 = pd.concat([df_2015,Year],axis=1)
```

```python
df_2015.head()
```

```python
Year = pd.Series([2016 for i in range(df_2016.shape[0])],name='Year')
df_2016 = pd.concat([df_2016,Year],axis=1)
```

```python
Year = pd.Series([2017 for i in range(df_2017.shape[0])],name='Year')
df_2017 = pd.concat([df_2017,Year],axis=1)
```

```python
df_2016.head()
```

```python
df_2017.head()
```

```python
df_2015.shape
```

```python
df_2016.shape
```

```python
df_2017.shape
```

```python
df_final = pd.concat([df_2015,df_2016,df_2017])
```

```python
df_final.head()
```

```python
df_final.sort_values(by='Country Name',inplace=True)
df_final = df_final.reset_index()
df_final.drop(columns=['index'],inplace=True)
```

```python
df1_2015 = pd.read_csv('../Datasets/Country_Happiness_Index/2015.csv')
df1_2016 = pd.read_csv('../Datasets/Country_Happiness_Index/2016.csv')
df1_2017 = pd.read_csv('../Datasets/Country_Happiness_Index/2017.csv')
```

```python
df1_2015.head()
```

```python
df1_2016.head()
```

```python
df1_2017.head()
```

```python
df1_2015 = df1_2015[['Country','Happiness Score']]
df1_2016 = df1_2016[['Country','Happiness Score']]
df1_2017 = df1_2017[['Country','Happiness.Score']]
df1_2017 = df1_2017.rename(columns={'Happiness.Score':'Happiness Score'})
```

```python
Year = pd.Series([2015 for i in range(df1_2015.shape[0])],name='Year')
df1_2015 = pd.concat([df1_2015,Year],axis=1)
```

```python
Year = pd.Series([2016 for i in range(df1_2016.shape[0])],name='Year')
df1_2016 = pd.concat([df1_2016,Year],axis=1)
```

```python
Year = pd.Series([2017 for i in range(df1_2017.shape[0])],name='Year')
df1_2017 = pd.concat([df1_2017,Year],axis=1)
```

```python
df1 = pd.concat([df1_2015,df1_2016,df1_2017])
```

```python
df1.sort_values(by=['Country','Year'],inplace=True)
df1 = df1.reset_index() 
df1.drop(columns=['index'],inplace=True)
```

```python
df1.rename(columns={'Country':'Country Name'},inplace=True)
```

```python
df1.head()
```
